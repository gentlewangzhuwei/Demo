## 数据挖掘与机器学习课程介绍

###一、课程目标：

资深的数据挖掘工程师应该达到由易到难的四个层级的水平：
1. 工具：学会使用某个挖掘工具（python）
2. 应用：使用工具结合实际数据去做项目（三次作业）
3. 理论：理解方法背后的理论原理
4. 实现：能实现一个经典的机器学习算法

对于学生来讲完全打通四个层次可能需要很长的时间，不过达到初级工作要求的1、2层次还是很快的。本课程会要求大部分学生能达到1、2层级，优秀学生要求达到3、4层次

要求学生：
- 有兴趣，有时间，有目标
- 不怕代码，不怕公式，不怕英文
- 必须进行补充阅读和练习
- 软件环境推荐是linux/mac（win同学建议安装虚拟机）


###二、上课形式：

1. 教师讲授：
教师通过四次课集中时间讲授，四个层次的知识会都有所涉及。
2. 学生自学和作业：
在每次课的间隔两周时间中，要求学生自学教师推荐的阅读文献。并完成作业。
3. 互动环节：
借鉴翻转课堂的思路，在完成作业后，在课堂上对全班讲解其项目过程和结果，并由教师点评打分。


###三、课程载体：

1. 使用工具：
选择python，原因是语法比较简单统一，而且在大数据以及深度学习等方面扩展模块丰富。

2. 主要教材：
[Python Machine Learning](http://www.amazon.com/Python-Machine-Learning-Sebastian-Raschka/dp/1783555130/ref=sr_1_1?ie=UTF8&qid=1451095363&sr=8-1&keywords=Python+Machine+Learning)
内容比较新，理论和代码平衡的非常好

3. 课件形式：
python notebook的形式，能将代码、输出以及讲解内容统一到一份web文档中。
http://ipython.org/notebook.html
学生的作业也要求是这种格式上交

###六、如何评分
六部分构成：
- 第一次作业是属于较基础的数据分析报告。
- 第二次作业是属于中等难度的数据挖掘项目。
- 第三次作业是属于于较高难度的数据挖掘项目。
- 第四次作业是加分项目，自愿完成上交。
- 每个学生需提交一个文献阅读报告。
- 如果有主动课堂回答问题或演讲表现优秀，也有相应加分。
