```python
fig, ax = plt.subplots(figsize=(8,5)) # 用figsize定义图的大小（当图片保存到磁盘时具有一定的大小和纵横比）

ax.plot(x, y1, lw=1.5, color="blue", label="$y(x)$") # 图例 label 输入 latex 数学符号和公式
ax.plot(x, y2, lw=1.5, color="red", label="$y'(x)$")
ax.plot(x, y3, lw=1.5, color="green", label="$y''(x)$")

ax.plot(x, np.zeros_like(x), lw=0.5, color="black")
ax.plot([-3.33, -3.33], [0, (-3.33)**3 + 5*(-3.33)**2 + 10], ls='--', lw=0.5, color="black")
ax.plot([0, 0], [0, 10], lw=0.5, ls='--', color="black")# 连接原点和[0,10]的线段
ax.plot([0], [10], lw=0.5, marker='o', color="blue") # 画点，o要小写
ax.plot([-3.33], [(-3.33)**3 + 5*(-3.33)**2 + 10], lw=0.5, marker='o', color="blue")

ax.set_ylim(-15, 40) # 显示y轴的上下限
ax.set_yticks([-10, 0, 10, 20, 30])# 画y轴的刻度和x轴的刻度
ax.set_xticks([-4, -2, 0, 2])

ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.legend(loc=0, ncol=3, fontsize=14, frameon=False); #loc 会自动设置 legend 位置
# 显示图例位置，0表示最佳，best。ncol表示图例要分几列。frameon表示是否需要图例的外框。
```

## 2.Plot types

+ 基本图形类型
    + 线图line
    + 条形图bar （表现离散数据）
    + 饼图pie
    + 直方图 histogram（画连续数据）
    + 散点图 scatter （两个因素的相关性)

```python
fignum = 0

def hide_labels(fig, ax):
    global fignum
    ax.set_xticks([])
    ax.set_yticks([])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.axis('tight')
    
    fignum += 1
#     fig.savefig("plot-types-%d.pdf" % fignum)
    
```

```python
x = np.linspace(-3, 3, 25)
y1 = x**3+ 3 * x**2 + 10
y2 = -1.5 * x**3 + 10*x**2 - 15
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

ax.plot(x, y1) # 默认是线图
ax.plot(x, y2)
hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

ax.step(x, y1) # step 是阶梯图
ax.step(x, y2)

hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
width = 6/50.0
ax.bar(x - width/2, y1, width=width, color="blue")
ax.bar(x + width/2, y2, width=width, color="green") # 错开两个图, 所以调整 width/2

hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
ax.fill_between(x, y1, y2) # y1和 y2之间的面积填起来

hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
ax.hist(y2, bins=30) # histogram 和 bar chart 不一样, bar chart是画 discrete 的, histogram 是画 cts
ax.hist(y1, bins=30)

hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

ax.errorbar(x, y2, yerr=y1, fmt='o-') #误差棒, 上下限

hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

ax.stem(x, y2, 'b', markerfmt='bs') # 把 y 轴的距离明显标出来
ax.stem(x, y1, 'r', markerfmt='ro')

hide_labels(fig, ax)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

x = np.linspace(0, 5, 50)

ax.scatter(x, -1 + x + 0.25 * x**2 + 2 * np.random.rand(len(x))) # 二次曲线+随机数, 观察两个变量关系
ax.scatter(x, np.sqrt(x) + 2 * np.random.rand(len(x)), color="green")

hide_labels(fig, ax)
```

## Line Properties


```python
x = np.linspace(-5, 5, 5)
y = np.ones_like(x)

def axes_settings(fig, ax, title, ymax):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(0, ymax+1)
    ax.set_title(title)
    
fig, axes = plt.subplots(1, 4, figsize=(16, 3))     

# Line width
linewidths = [0.5, 1.0, 2.0, 4.0]
for n, linewidth in enumerate(linewidths):
    axes[0].plot(x, y + n, color="blue", linewidth=linewidth)
axes_settings(fig, axes[0], "linewidth", len(linewidths))

# Line style
linestyles = ['-', '-.', ':']
for n, linestyle in enumerate(linestyles):
    axes[1].plot(x, y + n, color="blue", lw=2, linestyle=linestyle)
# custom dash style
line, = axes[1].plot(x, y + 3, color="blue", lw=2)
length1, gap1, length2, gap2 = 10, 7, 20, 7
line.set_dashes([length1, gap1, length2, gap2])
axes_settings(fig, axes[1], "linestypes", len(linestyles) + 1)

# marker types
markers = ['+', 'o', '*', 's', '.', '1', '2', '3', '4']
for n, marker in enumerate(markers):
    # lw = shorthand for linewidth, ls = shorthand for linestyle
    axes[2].plot(x, y + n, color="blue", lw=2, ls=':', marker=marker)
axes_settings(fig, axes[2], "markers", len(markers))

# marker size and color
markersizecolors = [(4, "white"), (8, "red"), (12, "yellow"), (16, "lightgreen")]
for n, (markersize, markerfacecolor) in enumerate(markersizecolors):
    axes[3].plot(x, y + n, color="blue", lw=1, ls='-',
                 marker='o', markersize=markersize,
                 markerfacecolor=markerfacecolor, markeredgewidth=2)
axes_settings(fig, axes[3], "marker size/color", len(markersizecolors))
```

注意区别

fig, **ax** = plt.subplots() 画布里只有一个axes。 fig 后跟 ax

fig, **axes** = plt.subplots(1, 4, figsize=(16, 3)) 画布里有多个axes。 fig 后跟 axes

### Legends

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

x = np.linspace(0, 1, 100)

for n in range(4): #用循环去画
    axes[n].plot(x, x, label="y(x) = x")
    axes[n].plot(x, x + x**2, label="y(x) = x + x**2")
    axes[n].legend(loc=n+1) #location 1,2,3,4就是4个角
    axes[n].set_title("legend(loc=%d)" % (n+1))

fig.tight_layout()
#fig.savefig("legend-loc.pdf")
```

### Twin axes

```python
fig, ax1 = plt.subplots(figsize=(8, 4))

r = np.linspace(0, 5, 100)
a = 4 * np.pi * r ** 2  # area
v = (4 * np.pi / 3) * r ** 3  # volume


ax1.set_title("surface area and volume of a sphere", fontsize=16)
ax1.set_xlabel("radius [m]", fontsize=16)

# 画表面积的曲线，并设置其y轴
ax1.plot(r, a, lw=2, color="blue")
ax1.set_ylabel("surface area ($m^2$)", fontsize=16, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

    
# 画体积的曲线，并设置其y轴
ax2 = ax1.twinx() #在前一个基础之上加轴
ax2.plot(r, v, lw=2, color="red")
ax2.set_ylabel("volume ($m^3$)", fontsize=16, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
    
fig.tight_layout()
fig.savefig("ch4-axis-twin-ax.pdf")
```

## 4.Advanced grid layout

### Inset

```python
fig = plt.figure(figsize=(8, 4))

def f(x):
    return 1/(1 + x**2) + 0.1/(1 + ((3 - x)/0.1)**2)

def plot_and_format_axes(ax, x, f, fontsize): #设置x轴和y轴的位置和轴的名称，字体大小
    ax.plot(x, f(x), linewidth=2)
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
    ax.set_xlabel(r"$x$", fontsize=fontsize)
    ax.set_ylabel(r"$f(x)$", fontsize=fontsize)
    
# main graph
ax = fig.add_axes([0.1, 0.15, 0.8, 0.8], axisbg="#f5f5f5") #左上角, 右上角的位置 # none则透明
x = np.linspace(-4, 14, 1000)
plot_and_format_axes(ax, x, f, 18)

# inset
x0, x1 = 2.5, 3.5
ax = fig.add_axes([0.5, 0.5, 0.38, 0.42], axisbg='none') #把嵌入的图画在右上角
x = np.linspace(x0, x1, 1000)
plot_and_format_axes(ax, x, f, 14)

#fig.savefig("ch4-advanced-axes-inset.pdf")
```

### Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(6, 6), sharex=True, sharey=True, squeeze=False)

x1 = np.random.randn(100)
x2 = np.random.randn(100)

axes[0, 0].set_title("Uncorrelated")
axes[0, 0].scatter(x1, x2)

axes[0, 1].set_title("Weakly positively correlated")
axes[0, 1].scatter(x1, x1 + x2)

axes[1, 0].set_title("Weakly negatively correlated")
axes[1, 0].scatter(x1, -x1 + x2)

axes[1, 1].set_title("Strongly correlated")
axes[1, 1].scatter(x1, x1 + 0.15 * x2)

axes[1, 1].set_xlabel("x")
axes[1, 0].set_xlabel("x")
axes[0, 0].set_ylabel("y")
axes[1, 0].set_ylabel("y")

plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95, wspace=0.1, hspace=0.2)

#fig.savefig("ch4-advanced-axes-subplots.pdf")
```
