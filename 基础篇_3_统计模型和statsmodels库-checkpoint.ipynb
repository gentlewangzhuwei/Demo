```python
def plot_rv_distribution(X, axes=None):
    ## """Plot the PDF, CDF, SF and PPF of a given random variable"""
    if axes is None:
        fig, axes = plt.subplots(1, 2, figsize=(8, 3))
    
    x_min_999, x_max_999 = X.interval(0.999)
    x999 = np.linspace(x_min_999, x_max_999, 1000)

    x_min_95, x_max_95 = X.interval(0.95)
    x95 = np.linspace(x_min_95, x_max_95, 1000)

    if hasattr(X.dist, 'pdf'): # 如果是 cts
        axes[0].plot(x999, X.pdf(x999), label="PDF")
        axes[0].fill_between(x95, X.pdf(x95), alpha=0.25)
    else: # 如果是 discrete
        x999_int = np.unique(x999.astype(int))
        axes[0].bar(x999_int, X.pmf(x999_int), label="PMF")
    axes[1].plot(x999, X.cdf(x999), label="CDF")
    
    for ax in axes:
        ax.legend()
    
    return axes
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 9))

X = stats.norm()
plot_rv_distribution(X, axes=axes[0, :]) 
axes[0, 0].set_ylabel("Normal dist.") #第一个图画标准正态分布
X = stats.f(2, 50)
plot_rv_distribution(X, axes=axes[1, :]) 
axes[1, 0].set_ylabel("F dist.") #第二个图画F分布
X = stats.poisson(5)
plot_rv_distribution(X, axes=axes[2, :])
axes[2, 0].set_ylabel("Poisson dist.") #第二个图画泊松分布
fig.tight_layout()
```

# 3.Hypothesis testing

## 显著性检验
显著性检验（significance test）就是事先对总体（随机变量）的参数或总体分布形式做出一个假设，然后利用样本信息来判断这个假设（备择假设）是否合理，即判断总体的真实情况与原假设是否有显著性差异。或者说，显著性检验要判断样本与我们对总体所做的假设之间的差异是纯属机会变异，还是由我们所做的假设与总体真实情况之间不一致所引起的。 显著性检验是针对我们对总体所做的假设做检验，其原理就是“小概率事件实际不可能性原理”来接受或否定假设。

步骤：
- 构造原假设（Null Hypothesis）和备择假设
- 选择某个检验统计量，e.g.T检验统计量
- 收集数据
- 根据数据计算统计量和相应的P值
- 若P值很小，则拒绝原假设

把检验统计量的样本观察值和临界值比较，或者把观察到的显著水平与显著水平标准比较；最后按检验规则作出检验决策。当样本值落入拒绝域时，表述成:“拒绝原假设”，“显著表明真实的差异存在”；当样本值落入接受域时，表述成：“没有充足的理由拒绝原假设”，“没有充足的理由表明真实的差异存在”。另外，在表述结论之后应当注明所用的显著水平。一个很通俗易懂的解释。

### T检验
t检验是用t分布理论来推论差异发生的概率，从而比较两个平均数的差异是否显著。它与z检验、卡方检验并列。
t检验分为单总体检验和双总体检验。
单总体t检验是检验一个样本平均数与一个已知的总体平均数的差异是否显著。当总体分布是正态分布，如总体标准差未知且样本容量小于30，那么样本平均数与总体平均数的离差统计量呈t分布。
双总体t检验是检验两个样本平均数与其各自所代表的总体的差异是否显著。双总体t检验又分为两种情况，一是独立样本t检验，一是配对样本t检验。

## 3. Straight-line Fit

```python
N = 50
m_true = 2
b_true = -1
dy = 2.0

np.random.seed(0)
xdata = 10*np.random.random(N)
ydata = np.random.normal(b_true + m_true * xdata, dy)# dy 相当于标准差

plt.errorbar(xdata, ydata, dy,
             fmt='.k', ecolor='lightgray');
```

### Method1 using ``scipy.optimize.minimize``

```python
def chi2(theta, x, y):
    return np.sum(((y - theta[0] - theta[1] * x)) ** 2)# 误差平方和，让误差平方和最小
```

```python
theta_guess = [0, 1]
theta_best = opt.minimize(chi2, theta_guess, args=(xdata, ydata)).x #使用minimize函数求解theta的值
print(theta_best)
```

```python
xfit = np.linspace(0, 10)
yfit = theta_best[0] + theta_best[1] * xfit # 预测值

plt.errorbar(xdata, ydata, dy,
             fmt='.k', ecolor='lightgray');
plt.plot(xfit, yfit, '-k');
```

## 1.Statistical models and patsy formula

### 线性模型基本概念

* 线性(回归)模型, 假设因变量 y 与一个或多个自变量 X 之间成线性关系, 用最小二乘法(OLS)来确定相应的系数
![](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/400px-Linear_regression.svg.png)

* $y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \epsilon$
    * y: dependent variable/endogenous variable
    * $x_i$: explanatory variables/independent variables/exogenous variables
    * 所有 x 构成的矩阵 X 称为设计矩阵(design matrix), 可以用 `np.vstack` 函数.
    * $\beta_i$: regression coefficients, $\beta_0$: intercept.
    * $\epsilon$: error term.

* OLS 的 python 实现
    * np.linalg.lstsq(X, y)
    * sm.OLS(y, X).fit()
    * smf.ols(formula, dataframe).fit()  → 当交互项很多时, 用这种方法比较方便
    * 具体到公式的形式, 可以用 `patsy.dmatrices` 方便的写出各种形式的公式, 比如 "y ~ a * b" 表示截距, a, b 与 a×b 四项, 还可以自定义公式, 还可以表示分类变量(category variables).
    
* statsmodels 模块可以实现多种最小二乘法, 主要有
    * [Ordinary Least Squares(OLS)](http://statsmodels.sourceforge.net/stable/examples/notebooks/generated/ols.html), 普通最小二乘法, 也是最常用的最小二乘法
    * [Weighted Least Squares(WLS)](http://statsmodels.sourceforge.net/stable/examples/notebooks/generated/wls.html), 加权最小二乘法
    * [Generalized Least Squares](http://statsmodels.sourceforge.net/stable/examples/notebooks/generated/gls.html), 广义最小二乘法

## 2. Linear regression
### 线性回归

* `model = smf.ols("y ~ x1 + x2", data)`
  
  `result = model.fit()`
  
* `result.summary()` 输出中, 主要项的解释
    * R-squared: 判定系数, [0, 1], 拟合的好坏
    * 各回归系数的值和p值: 判断该系数是否显著, 是否可以舍弃

* 残差(residual): 真实值和拟合值之差
    * result.resid

* 测试是否符合正态分布
    * stats.normaltest(通过 p 值来判断)
    * smg.qqplot(通过图像来判断, 应为45度直线)

* 在确定回归系数, 即确定线性回归方程后, 下一步要做的是预测.
    * result.predict

```python
import pandas as pd
np.random.seed(123456789)
N = 100 # sample size
x1 = np.random.randn(N)
x2 = np.random.randn(N)
data = pd.DataFrame({"x1": x1, "x2": x2})
```

```python
data["y_true"] = y_true(x1, x2)
```

```python
e = np.random.randn(N) #noise, ~Normal
```

```python
data["y"] = data["y_true"] + e #增加noise
```

## 3. logistic regression

### 逻辑回归

* y 为概率值.
* 逻辑函数的公式: $y = \frac {1}{1+e^{-(\beta_0 + \beta_1 x)}}$
* python 实现
    * `result = smf.logit(formula, data).fit()`
    * `result.summary()` 输出中, 主要关注的项
        * Pseudo R-squared
        * 各回归系数的值和 p 值
    * `result.get_margeff().summary()` 
    
* 决策边界, 线性分类器
* 是一种广义的线性模型.
* 不止两个分类值的逻辑回归
    * Andrew Ng 讲过 one vs all 方法.

$log\frac{p}{1-p} = \beta_0+\beta_1 x$

p 为概率, $\frac{p}{1-p}$  称为 odds

```python
df = sm.datasets.get_rdataset("iris").data
# 读取iris数据
#df = pd.read_csv('data/iris.csv')
```

```python
# 只取两种 species
df_subset = df[(df.Species == "versicolor") | (df.Species == "virginica" )].copy()
```

```python
# 映射
df_subset.Species = df_subset.Species.map({"versicolor": 1, "virginica": 0})
```

```python
# 点换成下划线, 避免出现问题
df_subset.rename(columns={"Sepal.Length": "Sepal_Length", "Sepal.Width": "Sepal_Width",
                          "Petal.Length": "Petal_Length", "Petal.Width": "Petal_Width"}, inplace=True)
```

```python
df_subset.head(3)
```

```python
# 预测种类, logit
model = smf.logit("Species ~ Sepal_Length + Sepal_Width + Petal_Length + Petal_Width", data=df_subset)
```

```python
result = model.fit()
```

```python
result.summary()
```

 看上面 coeff 的 p 值, 最后两个比较显著

**Note:** Sepal_Length and Sepal_Width do not seem to contribute much to predictiveness of the model. 

```python
# 改 model,去掉不显著的
model = smf.logit("Species ~ Petal_Length + Petal_Width", data=df_subset)
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
params = result.params
beta0 = (0.5-params['Intercept'])/params['Petal_Width']
beta1 = -params['Petal_Length']/params['Petal_Width']
# 决策边界 0.5=intercept+params['Petal_Length']*x1+params['Petal_Width']*x2-->x2=beta0+beta1*x1
```

```python
df_new = pd.DataFrame({"Petal_Length": np.random.randn(20)*0.5 + 5,
                       "Petal_Width": np.random.randn(20)*0.5 + 1.7})
```

```python
df_new["P-Species"] = result.predict(df_new)# predict 新数据
```

```python
df_new["P-Species"].head(3)#概率值
```

```python
df_new["Species"] = (df_new["P-Species"] > 0.5).astype(int)#0.5做临界点, 大于0.5或小于0.5为分类
```

```python
df_new.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

ax.plot(df_subset[df_subset.Species == 0].Petal_Length.values,
        df_subset[df_subset.Species == 0].Petal_Width.values, 's', label='virginica')
ax.plot(df_new[df_new.Species == 0].Petal_Length.values,
        df_new[df_new.Species == 0].Petal_Width.values,
        'o', markersize=10, color="steelblue", label='virginica (pred.)')

ax.plot(df_subset[df_subset.Species == 1].Petal_Length.values,
        df_subset[df_subset.Species == 1].Petal_Width.values, 's', label='versicolor')
ax.plot(df_new[df_new.Species == 1].Petal_Length.values,
        df_new[df_new.Species == 1].Petal_Width.values,
        'o', markersize=10, color="green", label='versicolor (pred.)')

_x = np.array([4.0, 6.1])
ax.plot(_x, beta0 + beta1 * _x, 'k')#决策线

ax.set_xlabel('Petal length')
ax.set_ylabel('Petal width')
ax.legend(loc=2)
fig.tight_layout()
# 线区分, 线性分类器
```
